Hierfür wurde ein Datensatz mit zwei Ordnern erstellt, einer mit gesunden Äpfeln und der andere mit defektbehafteten Äpfeln. Diese befinden sich unter: https://data.ips.biba.uni-bremen.de/Lehre/dsmlpl_SS21/Apples_Defect_Healthy.tgz

```python
from fastai.vision.all import *
import fastai
print(fastai.__version__ ) #soll > als 2.0 sein
```

```python
# Daten herunterladen und path speichern
url_photos = 'https://data.ips.biba.uni-bremen.de/Lehre/dsmlpl_SS21/Apples_Defect_Healthy.tgz'
path = untar_data(url_photos)
path.ls()
```

```python
# Bilder lesen
files = get_image_files(path)
len(files)
```

Nun können die Bilder betrachtet, ein Datablock erstellt und ein Batch angezeigt werden.

```python
# Ein Bild auswählen und zeigen
im = PILImage.create(files[6])
im.to_thumb(160)
```

## DataBlock

An sich ist ein DataBlock in Fastai nur eine Vorlage, um die Daten zusammenzustellen. 

Normalerweise wird der DataBlock aufgebaut, indem die Folgende Fragen beantwortet werden:

- Datentyp von Inputs und Outputs? Hier Bilder und Klassen
- Wo befinden sich die Daten? Hier im Dateinamen in Unterordnern
- Wie werden die Daten aufgeteilt? Hier zufällig
- Muss etwas auf die Inputs angewendet werden? Hier nein
- Muss etwas auf die Outputs angewandt werden? Hier die Funktion parent_label
- Müssen wir etwas auf geformte Elemente anwenden? Hier eine Größenänderung (Resize)
- Müssen wir etwas auf batch_transformations anwenden? Hier nichts

Damit kann ein Dataloader erstellt werden. Dieser erhält die Trainings- und Validationsdatensätze, um die CNN zu trainieren!

```python
# Data block erstellen
Apple_DB = DataBlock(blocks=(ImageBlock, CategoryBlock), 
                 get_items=get_image_files, 
                 splitter=RandomSplitter(0.2,seed=7),
                 get_y=parent_label ,
                 item_tfms=Resize(224),
                )
```

```python
# Bilder im Dataloader über den Datablock importieren
dls = Apple_DB.dataloaders(path)
```

```python
# Beispielbatch zeigen
dls.show_batch()
```

Zum automatisierten Erkennen der unversehrten und defekten Äpfel wird ein Neuronales Netz trainiert. Ein solches vortrainiertes wird zuerst Online abgerufen, anschließend wird eine passende Lernrate ermittelt und schließlich beginnt der eigentliche Trainingsvorgang.

```python
# learner Objekt
learn = cnn_learner(dls, resnet34, metrics=(error_rate,accuracy))
```

```python
# passendes learning rate finden
learn.lr_find()
```

```python
# tranieren
max_lr=1e-3
learn.fine_tune(5,max_lr,2)
```

Ist dieser Vorgang abgeschlossen, kann die Lernkurve einmal geplottet werden.

```python
learn.recorder.plot_loss()
```

```python
# Ergebnisse Zeigen
learn.show_results()
```

Um die Auswertung besser nachvollziehen zu können, werden im nächsten Schritt nicht nur die Vorhersage mit höhem Loss und auch die jeweiligen Wahrscheinlichkeiten angezeigt.

```python
# Ergebnisse verstehen
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(9, figsize=(15,10))
```

```python
interp.plot_confusion_matrix()
```
